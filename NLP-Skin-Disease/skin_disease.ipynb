```python
import scipy
from sentence_transformers import SentenceTransformer
```

```python
# Load the tokenizer
tokenizer = SentenceTransformer('sentence-transformers/all-MiniLM-L12-v2')
# tokenizer = SentenceTransformer('sentence-transformers/all-mpnet-base-v2')
```

```python
sentences=finaldata['combined'].tolist()

# sentences
```

```python
text_embeddings=tokenizer.encode(sentences)
```

```python
text_embeddings
```

```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS

nlp = spacy.load('en_core_web_sm')
```

```python
# define the preprocess function
def preprocess_symptom(symptom_text):
    sym = nlp(symptom_text)
    cleaned_sym = []
    for words in sym:
        if words.text.lower() not in STOP_WORDS and words.text.isalpha():
            cleaned_sym.append(words.lemma_.lower())
    return ' '.join(cleaned_sym)
```

```python
query = "My scalp is very itchy, and I see flakes in my hair. What could be causing this?"
```

```python
query_embeddings = tokenizer.encode(query)
```

```python
# query_embeddings
```

```python
query_embeddings=query_embeddings.reshape(1,-1)
```

```python
# query_embeddings
```

```python
#cosine similarity
distances_=scipy.spatial.distance.cdist(text_embeddings, query_embeddings,"cosine")
distances=1-distances_
# distances
```

```python
import numpy as np

match = np.sort(distances.flatten())[::-1]
```

```python
match
```

```python
match[0]
```

```python
findindex = list(distances.flatten())
index1 = findindex.index(match[0])
index2= findindex.index(match[1])
index3= findindex.index(match[2])
```

```python
sentences[index3]
```

```python
confidence = 1-distances_.flatten()
```

### symptoms for query

```python
# Load JSON data from file
with open('all_symptoms.json', 'r') as f:
    symptom_data = json.load(f)

#view the data
symptom_data
```

```python
# Preprocess all symptoms and store them in a list
preprocessed_symptoms = [preprocess_symptom(symptom) for symptom in symptom_data]
for original, preprocessed in zip(symptom_data, preprocessed_symptoms):
    print(f"Original: {original} -> Preprocessed: {preprocessed}")
```

```python
# preprocessed_symptoms
```

```python
# Convert query to lower case for matching
query_lower = query.lower()
query_lower = preprocess_symptom(query)
# Find matched symptoms in the query
matched_symptoms = set()
for sym in preprocessed_symptoms:
    if sym in query_lower:
        matched_symptoms.add(sym.strip())

# Convert matched_symptoms to a sorted list and join them to a string
identified_symptoms = ', '.join(sorted(matched_symptoms))
```

```python
responses = [{'identified symptoms': identified_symptoms}]
response = np.array(sentences)[[index1,index2,index3]]
confidence_ = confidence[[index1,index2,index3]]
for mconfidence,outputtext_ in zip(confidence_,response):
    outputtext= outputtext_.split('|')
    disease = outputtext[0].split(':')[-1]
    symptom = outputtext[1].split(':')[-1]
    etiology = outputtext[2].split(':')[-1]

    finalresponse = {
    'disease':disease,
    'symptom':symptom,
    'etiology': etiology,
    'confidence':mconfidence
    }
    responses.append(finalresponse)


```
